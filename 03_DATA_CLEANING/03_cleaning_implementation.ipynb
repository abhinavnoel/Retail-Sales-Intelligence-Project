## Step 2 — Data Type Conversions


This section performs all dtype corrections (date, numeric, and coerced values), following the Cleaning Blueprint.


```python
df = pd.read_csv(r"E:\M90\PROJECTS\MY PROJECTS\PROJECT 3 — Retail Store Sales Intelligence & Forecasting Engine\03_DATA_CLEANING\01_raw_snapshot.csv")
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
```

```python
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
```

```python
df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce')
```

```python
df.to_csv(r"E:\M90\PROJECTS\MY PROJECTS\PROJECT 3 — Retail Store Sales Intelligence & Forecasting Engine\03_DATA_CLEANING\clean_step1_dtypes.csv", index=False)
```

### Data Type Conversion Summary
- Date converted to datetime.
- Amount and Qty converted to numeric.
- Invalid values coerced to NaN.
- Saved as clean_step1_dtypes.csv.


```python
# Apply basic text cleanup to all object columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip().str.lower()
```

```python
df['promotion-ids'] = df['promotion-ids'].astype(str).str.strip().str.replace(' ,', ',', regex=False).str.replace(', ', ',', regex=False)
```

```python
df.to_csv(r"E:\M90\PROJECTS\MY PROJECTS\PROJECT 3 — Retail Store Sales Intelligence & Forecasting Engine\03_DATA_CLEANING\clean_step2_text.csv", index=False)
```

### Text Normalization Summary
- Trimmed whitespace across all text columns.
- Converted text to lowercase for consistency.
- Standardized basic formatting in promotion-ids.
- Saved as clean_step2_text.csv.


```python
# Clean Amount
df['Amount'] = df['Amount'].astype(str).str.replace(',', '', regex=False).str.replace('₹', '', regex=False)
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
```

```python
# Clean Quantity
df['Qty'] = df['Qty'].astype(str).str.replace(',', '', regex=False)
df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce')
```

```python
# Clean shipping charge if it exists
if 'ship-charge' in df.columns:
    df['ship-charge'] = df['ship-charge'].astype(str).str.replace(',', '', regex=False).str.replace('₹', '', regex=False)
    df['ship-charge'] = pd.to_numeric(df['ship-charge'], errors='coerce')
```

```python
df.to_csv(r"E:\M90\PROJECTS\MY PROJECTS\PROJECT 3 — Retail Store Sales Intelligence & Forecasting Engine\03_DATA_CLEANING\clean_step3_numeric.csv", index=False)
```

### Numeric Cleaning Summary
- Removed commas and currency symbols from Amount, Qty, and any shipping charge fields.
- Converted these fields to numeric.
- Invalid values were safely coerced to NaN.
- Saved as clean_step3_numeric.csv.


```python
# Ensure promotion-ids is string
df['promotion-ids'] = df['promotion-ids'].astype(str).str.strip()
```

```python
# promo_flag: 1 if promotion exists, else 0
df['promo_flag'] = df['promotion-ids'].apply(lambda x: 0 if x == 'nan' or x == '' else 1)
```

```python
# promo_count: count number of promotions (by counting commas + 1)
df['promo_count'] = df['promotion-ids'].apply(
    lambda x: 0 if x == 'nan' or x == '' else len(x.split(',')))
```

```python
df.to_csv(r"E:\M90\PROJECTS\MY PROJECTS\PROJECT 3 — Retail Store Sales Intelligence & Forecasting Engine\03_DATA_CLEANING\clean_step4_promo.csv", index=False)
```

### Promotion Parsing Summary
- Created promo_flag (1 if promotion applied, else 0).
- Created promo_count (number of promotions based on comma-separated values).
- Kept promotion-ids in cleaned format for reference.
- Saved as clean_step4_promo.csv.


```python
# Normalization: lowercase + strip (already done earlier), but create clean columns

df['status_clean'] = df['Status'].astype(str).str.strip().str.lower()
df['fulfilment_normalized'] = df['Fulfilment'].astype(str).str.strip().str.lower()
df['courier_status_normalized'] = df['Courier Status'].astype(str).str.strip().str.lower()
df['category_clean'] = df['Category'].astype(str).str.strip().str.lower()
```

### Category Normalization Summary
- Created clean versions of Status, Fulfilment, Courier Status, and Category.
- Normalization is simple: lowercase + strip.
- Saved as clean_step5_category.csv.


```python
df['return_flag'] = df['status_clean'].str.contains('return', na=False).astype(int)
df['cancel_flag'] = df['status_clean'].str.contains('cancel', na=False).astype(int)
```

```python
df.to_csv(r"E:\M90\PROJECTS\MY PROJECTS\PROJECT 3 — Retail Store Sales Intelligence & Forecasting Engine\03_DATA_CLEANING\clean_step6_derived.csv", index=False)
```

### Derived Flags Summary
- Created return_flag: 1 if status contains 'return'.
- Created cancel_flag: 1 if status contains 'cancel'.
- Saved as clean_step6_derived.csv.


```python
#  trim + lowercase for city/state/country
df['ship-city'] = df['ship-city'].astype(str).str.strip().str.lower()
df['ship-state'] = df['ship-state'].astype(str).str.strip().str.lower()
df['ship-country'] = df['ship-country'].astype(str).str.strip().str.lower()

# Postal code: trim and keep as string
df['ship-postal-code'] = df['ship-postal-code'].astype(str).str.strip()
```
